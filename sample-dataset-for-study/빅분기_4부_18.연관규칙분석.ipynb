```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np 
import pandas as pd 
```

# 1. 분석데이터 불러오기 및 확인

```python
data = pd.read_csv('Market_Basket.csv', header = None)
data.head()
```

```python
transactions = []
for i in range(data.shape[0]):
    transactions.append([str(data[j][i]) 
                         for j in range(data.shape[1]-data.isnull().sum(axis=1)[i])])
```

```python
transactions
```

# 2. apriori 모델 적용

```python
!pip install apyori
```

```python
from apyori import apriori
```

```python
rules = apriori(transactions, min_support = 0.015, min_confidence = 0.2,  
                min_lift = 1, min_length = 1)
```

```python
results = list(rules) 
```

```python
results
```

```python
print(len(results))
```

```python
df=pd.DataFrame(results)
df
```

```python
df.to_csv('apriori_result.csv')
```

# 3. 연관품목의 시각화

```python
ar=(df.iloc[1:74]['items'])
```

```python
ar
```

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(9,6))

G = nx.Graph()
G.add_edges_from(ar)
pos = nx.spring_layout(G)
nx.draw(G, pos, font_size=16, with_labels=False, edge_color='green',
        node_size=800, node_color=['red','green','blue','cyan','orange','magenta'])
for p in pos:  
    pos[p][1] += 0.07
nx.draw_networkx_labels(G, pos)
plt.show()
```
